###About the data being analyzed
####  This data is provided by the Consumer Financial Protection Bureau. Each week CFPB sends thousands of consumers’ complaints about financial products and services to companies for response. Complaints are listed in the database after the company responds or after they’ve had the complaint for 15 calendar days, whichever comes first. CFPB publishes the consumer’s description of what happened if the consumer opts to share it and after taking steps to remove personal information. The Consumer Complaint Database is a collection of 551,694 complaints, on a range of consumer financial products and services, sent to nearly 3,000 companies for response.

#### Prepare to collect accuracy and confusion matrices by models

```python
res=dict()
def init_res(d):
    d.clear()
def col_res(k,v,d):
    d[k]=v
def disp_res(d):
    for k,v in d.iteritems():
        print(k + ' is '+ str(v))
    
init_res(res) 
col_res('shape','red',res)
disp_res(res)
init_res(res)
print(res)
```

#### Logistic classifier
##### * Use logistic regression to classify the data set to evaluate timely response as the response characteristic using the Product, Issue, Company, Zip code and submiited via factors.*

#### Split into test and train sets

```python
train_ccsf,test_ccsf = ccsf.random_split(.8, seed=0)
```

#### Run logistic regression
##### *The probability of a timely response is modeled as a logistic function of Product, Issue, Company, Zip code and Submiited via features. *

#####  Logistic regression can 
###### (a) calculate classes based on probability of 0.5; OR
###### (b) determine an estimate of probability for a positive response OR
###### (c) distance from each response to the calculated linear decision boundary .
###### model.predict(test_data, output_type = "class" or"probability" or"margin")          

```python
trlc_model = graphlab.logistic_classifier.create(train_ccsf,
                                                     target='Timely response?',
                                                     features=['Product', 'Issue', 'Company','ZIP code',"Submitted via"],
                                                     validation_set=test_ccsf)
```

```python
trlc_model['coefficients']
```

#### Making predictions and evaluating results

##### *Estimates of the probability of a positive timely response is stored into the array*

```python
cl_trlc = trlc_model.classify(test_ccsf)
```

##### *Print the predicted class value with the associated probability*

```python
print cl_trlc
```

#### *What does the model predict and how does it compare?*

```python
trlc_result=trlc_model.evaluate(test_ccsf)
```

```python
print("Accuracy              :%s" %trlc_result['accuracy'])
print("Confusion Matrix      :%s" %trlc_result['confusion_matrix'])
```

```python
col_res('Logistic_accuracy',trlc_result['accuracy'],res)
#col_res('Logistic_accuracy',trlc_result['confusion_matrix'],res)
```

##### *Utilize the model and store the results in the form of a dictionary*

```python
trlc_results=trlc_model.evaluate(test_ccsf,metric='roc_curve')
```

```python
trlc_results
```

```python
trlc_model.show(view="Evaluation")
```

```python
trlc_model.evaluate(test_ccsf,metric='accuracy')
```

##### *Finally we will explore the idea that in our dataset the lack of timely response is unde-represented and therefore our model might not provide optimal results. We will go with a balanced model.*

```python
trlc_b_model=graphlab.logistic_classifier.create(train_ccsf,
                                                     target='Timely response?',
                                                     features=['Product', 'Issue', 'Company','ZIP code',"Submitted via"],
                                                     class_weights="auto",
                                                     validation_set=test_ccsf)
```

#### *What does the balanced model predict and how does it compare?*
##### *The balanced model does not seem to provide any improvements over the existing model.*

```python
trlc_b_result=trlc_model.evaluate(test_ccsf)
```

### Decision Tree models
#### * The boosted tree model uses multiple base models and combines results from these base models to make predictions. The base models are combined using gradient boosting.*
#### * Gradient boosting can model non-lnear interactions between the response and feature variables.*

```python
train_dt_ccsf,test_dt_ccsf = ccsf.random_split(.8, seed=0)
```

```python
trlc_dt_model = graphlab.boosted_trees_classifier.create(train_dt_ccsf,
                                                     target='Timely response?',
                                                     features=['Product', 'Issue', 'Company','ZIP code',"Submitted via"],
                                                     validation_set=test_dt_ccsf)
```

#### *What does the boosted model predict and how does it compare?*

```python
trlc_dt_result=trlc_dt_model.evaluate(test_dt_ccsf)
```
